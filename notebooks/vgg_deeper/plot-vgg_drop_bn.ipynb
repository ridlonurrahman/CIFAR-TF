```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

batch_size = 128
train_dir_d = '../vgg_dropout/cifar10_vgg_model/'
train_dir_bn = '../vgg_dropout_bn/cifar10_vgg_model/'
train_dir_v = '../vgg/cifar10_vgg_model/'
```

```python
training_data_v = np.genfromtxt (train_dir_v+'training_data.csv', delimiter=",")
validation_data_v = np.genfromtxt (train_dir_v+'validation_data.csv', delimiter=",")
validation_data_v = np.array(map(lambda x: [((x[0]-x[0]%10)/float(50000/float(batch_size))),x[1],x[2]],validation_data_v))
validation_data_v = validation_data_v[validation_data_v[:,0]<=60]
training_data_v = np.array(map(lambda x: [x[0]/(50000/float(batch_size)),x[1],x[2]], training_data_v))
training_v_df = pd.DataFrame(training_data_v, index=training_data_v.transpose()[0],columns=['step','error','accuracy'])
validation_v_df = pd.DataFrame(validation_data_v, index=validation_data_v.transpose()[0],columns=['step','error','accuracy'])
plot_v_df = pd.merge(training_v_df,validation_v_df,how='inner',on=['step','step'])
plot_v_data = plot_v_df.as_matrix().transpose()
```

```python
training_data_d = np.genfromtxt (train_dir_d+'training_data.csv', delimiter=",")
validation_data_d = np.genfromtxt (train_dir_d+'validation_data.csv', delimiter=",")
validation_data_d = np.array(map(lambda x: [((x[0]-x[0]%10)/float(50000/float(batch_size))),x[1],x[2]],validation_data_d))
validation_data_d = validation_data_d[validation_data_d[:,0]<=60]
training_data_d = np.array(map(lambda x: [x[0]/(50000/float(batch_size)),x[1],x[2]], training_data_d))
training_d_df = pd.DataFrame(training_data_d, index=training_data_d.transpose()[0],columns=['step','error','accuracy'])
validation_d_df = pd.DataFrame(validation_data_d, index=validation_data_d.transpose()[0],columns=['step','error','accuracy'])
plot_d_df = pd.merge(training_d_df,validation_d_df,how='inner',on=['step','step'])
plot_d_data = plot_d_df.as_matrix().transpose()
```

```python
training_data_bn = np.genfromtxt (train_dir_bn+'training_data.csv', delimiter=",")
validation_data_bn = np.genfromtxt (train_dir_bn+'validation_data.csv', delimiter=",")
validation_data_bn = np.array(map(lambda x: [((x[0]-x[0]%10)/float(50000/float(batch_size))),x[1],x[2]],validation_data_bn))
validation_data_bn = validation_data_bn[validation_data_bn[:,0]<=60]
training_data_bn = np.array(map(lambda x: [x[0]/(50000/float(batch_size)),x[1],x[2]], training_data_bn))
training_bn_df = pd.DataFrame(training_data_bn, index=training_data_bn.transpose()[0],columns=['step','error','accuracy'])
validation_bn_df = pd.DataFrame(validation_data_bn, index=validation_data_bn.transpose()[0],columns=['step','error','accuracy'])
plot_bn_df = pd.merge(training_bn_df,validation_bn_df,how='inner',on=['step','step'])
plot_bn_data = plot_bn_df.as_matrix().transpose()
```

```python
ax = plt.figure()
plt.figure(figsize=(15,10))

plt.subplot(221)
plt.plot(plot_v_data[0], plot_v_data[1], 'r-', label='VGGNet (training)')
plt.plot(plot_d_data[0], plot_d_data[1], 'b-', label='VGGNet + dropout (training)')
plt.plot(plot_bn_data[0], plot_bn_data[1], 'g-', label='VGGNet + dropout + BN (training)')
plt.title('Error training')
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend(loc='upper right', fontsize='medium')

plt.subplot(222)
plt.plot(plot_v_data[0], plot_v_data[2], 'r-', label='VGGNet (training)')
plt.plot(plot_d_data[0], plot_d_data[2], 'b-', label='VGGNet + dropout (training)')
plt.plot(plot_bn_data[0], plot_bn_data[2], 'g-', label='VGGNet + dropout + BN (training)')
'''plt.plot(acc_train_tanh.keys(), acc_train_tanh.values(), 'b-', label='tanh (training)')
plt.plot(acc_valid_tanh.keys(), acc_valid_tanh.values(), 'b--', label='tanh (validation)')
plt.plot(acc_train_relu.keys(), acc_train_relu.values(), 'g-', label='relu (training)')
plt.plot(acc_valid_relu.keys(), acc_valid_relu.values(), 'g--', label='relu (validation)')'''
plt.title('Accuracy training')
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right', fontsize='medium')

plt.subplot(223)
plt.plot(plot_v_data[0], plot_v_data[3], 'r-', label='VGGNet (validation)')
plt.plot(plot_d_data[0], plot_d_data[3], 'b-', label='VGGNet + dropout (validation)')
plt.plot(plot_bn_data[0], plot_bn_data[3], 'g-', label='VGGNet + dropout + BN (validation)')
plt.title('Error validation')
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.legend(loc='upper right', fontsize='medium')

plt.subplot(224)
plt.plot(plot_v_data[0], plot_v_data[4], 'r-', label='VGGNet (validation)')
plt.plot(plot_d_data[0], plot_d_data[4], 'b-', label='VGGNet + dropout (validation)')
plt.plot(plot_bn_data[0], plot_bn_data[4], 'g-', label='VGGNet + dropout + BN (validation)')
plt.title('Accuracy validation')
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.yticks(np.arange(0, 1.2, 0.2))
plt.legend(loc='lower right', fontsize='medium')

plt.savefig('vgg_dropout_bn.pdf', bbox_inches='tight')
plt.show()
```
